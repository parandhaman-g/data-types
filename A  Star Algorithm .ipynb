```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def dist(a,b):
    (x1,y1)=a
    (x2,y2)=b
    return((x1-x2)**2+(y1-y2)**2)**0.5
```

```python
g=nx.grid_graph(dim=[6,7])
nx.set_edge_attributes(g,{e:e[1][0]*2 for e in g.edges()},"cost")
```

```python
pos=nx.spring_layout(g)
nx.draw(g,pos,with_labels=True,node_color="#f16e44")
edge_labels=nx.get_edge_attributes(g,"cost")
nx.draw_networkx_edge_labels(g,pos,edge_labels=edge_labels)
plt.show()
```

```python
path=nx.astar_path(g,(0,0),(6,4),heuristic=dist,weight="cost")
```

```python
length=nx.astar_path_length(g,(0,0),(6,4),heuristic=dist,weight="cost")
```

```python
print('path:',path)
```

```python
print('path length:',length)
```
